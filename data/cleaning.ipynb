```python
# import libraries
import pandas as pd
import numpy as np
import re
```

```python
# import data
df = pd.read_csv("raw_data.csv")
df.head()
```

```python
# city (split into city and province)
def get_city(address):
    return address.split(',')[0]

df['city'] = df['address'].apply(lambda x: f"{get_city(x)}") 

def get_province(address):
    return address.split(',')[1].split(' ')[1] 

df['province'] = df['address'].apply(lambda x: f"{get_province(x)}")

df.head()
```

```python
# convert financial_transparency to x = 0, check = 1, check plus = 2
mapping = {'✖': 0, '✔+': 2, '✔': 1}
df['financial_transparency'] = df['financial_transparency'].replace(mapping)

df.head()
```

```python
# split contact into website and tel

df['contact'] = df['contact'].str.replace('\n\n', ' ')

def get_website(contact):
    return contact.split('Tel')[0]

df['website'] = df['contact'].apply(lambda x: f"{get_website(x)}") 
df['website'] = df['website'].str.replace('Website: ', '')

def get_tel(contact):
    parts = contact.split('Tel')
    if len(parts) >= 2:
        return parts[1].strip()
    else:
        return None

df['Tel'] = df['contact'].apply(get_tel)
df['Tel'] = df['Tel'].str.replace(':', '')


df.head()
```

```python
# Split category into main and sub categories
def get_category(cat):
    return cat.split(' - ')[0]

df['main_category'] = df['category'].apply(lambda x: f"{get_category(x)}") 

def get_subcat(cat):
    parts = cat.split(' - ')
    if len(parts) >= 2:
        return parts[1].strip()
    else:
        return None

df['sub_category'] = df['category'].apply(get_subcat)

df.head()
```

```python
# Drop original title in overview
def get_overview(overview):
    return overview.split(':\n\n')[1]

df['new_overview'] = df['overview'].apply(lambda x: f"{get_overview(x)}")

df['new_overview'] = df['new_overview'].str.replace('\n', ' ')

df.head()
```

```python
# drop original title in results and impacts
def get_results(res):
    parts = res.split('\n')
    if len(parts) >= 2:
        return parts[1]
    return None

df['new_results_and_impact'] = df['results_and_impact'].apply(lambda x: f"{get_results(x)}")


df.head()
```

```python
# Drop old columns
columns_to_drop = ['address', 'category', 'contact']
df.drop(columns_to_drop, axis=1, inplace=True)

df.head()
```

```python
csv_filename = 'clean_data.csv'
df.to_csv(csv_filename, index=False)

print("DataFrame converted to CSV and saved successfully.")
```
